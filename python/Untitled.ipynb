```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

```python
img = cv2.imread('video/image_05.png')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

```python
import os
image_name_list = os.listdir("video/")
image_name_list.sort()
image_dir_list = []
img_list = []
for image_name in image_name_list:
    # print("name = ",image_name)
    image_name = "video/"+image_name
    image_dir_list.append(image_name)
    img = cv2.imread(image_name)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR);
    img_list.append(img)

# show the first and the last
# plt.imshow(img_list[0])
plt.imshow(img_list[-1])
```

```python

# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from moviepy.editor import ImageSequenceClip
from IPython.display import HTML
```

```python
# write image sequence into a video 
# following https://zulko.github.io/moviepy/getting_started/videoclips.html
# video_clip = ImageSequenceClip(img_list, fps=25)
video_clip = ImageSequenceClip(image_dir_list, fps=10)
```

```python
output_clip = 'output.mp4'
%time video_clip.write_videofile(output_clip, audio=False)
```

```python
HTML("""
<video width="960" height="540" controls>
  <source src="{0}">
</video>
""".format(output_clip))
```
